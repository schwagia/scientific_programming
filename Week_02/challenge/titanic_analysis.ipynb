```python
import os
import pandas as pd

# Titanic-Datensatz laden
df = pd.read_csv("Titanic-Dataset.csv")

# Geschlecht in binäre Werte umwandeln
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})

# Subset nach den gegebenen Bedingungen erstellen
subset = df[(df["Survived"] == 1) & ((df["Sex"] == 1) & (df["Age"] > 45) | (df["Sex"] == 0) & (df["Age"] < 20))]

# Anzahl der Passagiere im Subset anzeigen
print("Anzahl der gefilterten Passagiere:", subset.shape[0])
```
